```python
import pandas as pd
df=pd.read_csv("Netflix_Engagement_Dataset.csv")
```

```python
print(df.dtypes)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

categ_cols = [
    "Device Used Most Often", "Genre Preference", "Region",
    "Payment History (On-Time/Delayed)", "Subscription Plan",
    "Promotional Offers Used"
]

encoders = {}  

for col in categ_cols:
    le = LabelEncoder()
    df[col] = df[col].astype(str)
    df[col] = le.fit_transform(df[col])
    encoders[col] = le  
```

```python
print(df.dtypes)
```

```python
def predict_churn(new_data, encoders, scaler, categ_cols, model):
    new_df = pd.DataFrame([new_data])

    for col in categ_cols:
        le = encoders[col]
        class_mapping = {label: idx for idx, label in enumerate(le.classes_)}

        val = new_df[col][0]
        if val in class_mapping:
            new_df[col] = class_mapping[val]
        else:
     
            new_df[col ] = -1

    new_df_scaled = scaler.transform(new_df)
    prediction = model.predict(new_df_scaled)

    return "Churn" if prediction[0] == 1 else "Not Churn"
```

```python
new_customer = {
    "Subscription Length (Months)": 12,
    "Customer Satisfaction Score (1-10)": 6,
    "Daily Watch Time (Hours)": 4,
    "Engagement Rate (1-10)": 7,
     "Device Used Most Often": "Mobile",
    "Genre Preference": "Romance",
    "Region": "North America",
    "Payment History (On-Time/Delayed)": "On-Time",  
    "Subscription Plan": "Standard",
    "Support Queries Logged": 1,
    "Age": 25,
    "Monthly Income ($)": 5000,
    "Promotional Offers Used": "Yes",
    "Number of Profiles Created": 3
}
print("Predicted Churn Status:", predict_churn(new_customer,encoders, scaler, categ_cols, model))
```
